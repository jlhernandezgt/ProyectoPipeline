```python
dataset = pd.read_csv('train_output.csv')
dataset.head(10)
```

```python
X = dataset.drop(['gender'], axis=1)
y = dataset['gender']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True, random_state=2022)
```

```python
modelo, scores, resultados = mf.train_model(X_train, y_train)
with open('tiempo_ejecucion.txt', 'w') as archivo:
    archivo.write(str(modelo))
```

```python
DT = DecisionTreeClassifier(random_state=2022)
grid_search_dt = {'max_depth': [2],'min_samples_leaf': [5],'criterion': ["gini"]}
gs_DT = GridSearchCV(estimator=DT, param_grid=grid_search_dt, cv=10, n_jobs=-1, verbose=1, scoring = "accuracy")
gs_DT.fit(X_train, y_train)
```

```python
gs_DT.predict(X_test)
```

```python
y_train_pred = gs_DT.predict(X_train)
y_test_pred = gs_DT.predict(X_test)
```

```python
conf_matrix = pd.crosstab(y_test, y_test_pred, rownames=["observación"], colnames=["Predicción"])
print("Matriz de Confusión: \n\n", conf_matrix)
```

```python
fn.validacion_dt(conf_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, gs_DT.predict(X_test)))
```
